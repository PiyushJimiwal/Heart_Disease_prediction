```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
import seaborn as sns

# --- loading data ---

def load_data(filepath):
    """
    Loads a CSV file into a pandas DataFrame

    Prints out basic info about the dataset

    Returns the DataFrame if successful, otherwise returns None
    """
    try:
        df = pd.read_csv(filepath)
        print("Dataset successfully loaded.")
        print(f"Shape of the data: {df.shape}")
        print("\n Information about dataset:")
        df.info()
        print("\nLoad Head of dataset:")
        print(df.head())
        return df
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}. Please make sure the file path is correct.")
        return None

def check_missing_values(df):
    """
    Checks and displays the number of missing values per column."

    Args:
    df (pd.DataFrame) – The DataFrame to be checked.

    Returns:
    pd.Series – A Series containing the number of missing values per column.

    """
    print("\nCheck for missing values:")
    missing_values = df.isnull().sum()
    # Filter only the columns with missing values.
    missing_values = missing_values[missing_values > 0]
    if missing_values.empty:
        print("There is no missing values.")
    else:
        print(missing_values)
    return missing_values

def explore_target_distribution(df, target_column):
    """
    Explore the distribution of the target column and visualize it..

    Args:
        df (pd.DataFrame): DataFrame.
        target_column (str): The name of the column that repersent your label.
    """
    print(f"\nCount of each class in target '{target_column}':")
    print(df[target_column].value_counts())
    # normalize give the precentage instead of columns
    print(f"Proportion of target 1 (Heart Disease): {df[target_column].value_counts(normalize=True)[1]:.2%}")

    plt.figure(figsize=(6, 4))
    sns.countplot(x=target_column, data=df)
    plt.title(f'Distribution of Target column ({target_column})')
    plt.xlabel('Heart Disease (0: No, 1: Yes)')
    plt.ylabel('Number of Patients')
    plt.show()

# --- Cleaning Data ---

def clean_data(df):
    """
    Cleaning the data, Handling values that are '?' Convert to numeric,
    Fill missing values with median

    Args:
        df (pd.DataFrame): DataFrame to be checked.

    Returns:
        pd.DataFrame: DataFrame after cleaning.
    """
    df_cleaned = df.copy()

    # Replace '?' with NaN in the relevant columns
    cols_to_check = ['ca', 'thal']
    for col in cols_to_check:
        if col in df_cleaned.columns:
            # Replacing ? to nan
            df_cleaned[col] = df_cleaned[col].replace('?', np.nan)

    #Convert all columns to numeric type, forcing any errors to become NaN.
    for col in df_cleaned.columns:
        # The parameter errors='coerce' tells pandas:
        # “If you can’t convert a value to a number,
        # don’t throw an error — instead, turn it into NaN (missing value).”
        df_cleaned[col] = pd.to_numeric(df_cleaned[col], errors='coerce')

    # Impute missing values (NaN) with the median of each column
    print("\nHandling missing values (if any) using the median...")
    for col in df_cleaned.columns:
        if df_cleaned[col].isnull().sum() > 0:
            median_val = df_cleaned[col].median()
            df_cleaned[col].fillna(median_val, inplace=True)
            print(f"Column {col} was filled with its median value:: {median_val}")

    print("\nCheck for missing values after cleaning.")
    missing_after_clean = df_cleaned.isnull().sum()[df_cleaned.isnull().sum() > 0]
    if missing_after_clean.empty:
        print("All missing values have been handled.")
    else:
        print(missing_after_clean)

    return df_cleaned

# --- Pre-Processeing Features---

def preprocess_features(df, target_column):
    """
    Separates the features (X) and target (y), and identifies numerical and categorical columns.

    Args:
          df (pd.DataFrame): The DataFrame to be processed.
          target_column (str): The name of the target column.

    Returns:
            tuple: (X, y, preprocessor, numerical_cols, categorical_cols)
            - X (pd.DataFrame): The feature set.
            - y (pd.Series): The target variable.
            - preprocessor (ColumnTransformer): The sklearn preprocessing object.
            - numerical_cols (list): List of numerical column names.
            - categorical_cols (list): List of categorical column names.
    """
    X = df.drop(columns=[target_column])
    y = df[target_column]

    # Identify columns based on UCI Cleveland dataset description
    numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
    categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']

    #  checks if each listed column actually exists in the dataset.
    numerical_cols = [col for col in numerical_cols if col in X.columns]
    categorical_cols = [col for col in categorical_cols if col in X.columns]

    # Create preprocessor for feature transformation
    # Use StandardScaler for numerical features
    # Use OneHotEncoder for categorical features
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), numerical_cols),
            ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
        ],
        remainder='passthrough'
    )

    print(f"\nNumerical columns to be scaled: {numerical_cols}")
    print(f"Categorical columns to be encoded: {categorical_cols}")

    return X, y, preprocessor, numerical_cols, categorical_cols

# --- split Data ---

def split_data(X, y, test_size=0.2, random_state=42):
    """
    Splits the data into training and testing sets.
    Uses stratify=y to ensure the same class distribution in both sets.

    Args:
        X (pd.DataFrame): Features.
        y (pd.Series): Target.
        test_size (float): Proportion of data to use for testing (default: 0.2).
        random_state (int): Seed for reproducible shuffling (default: 42).

    Returns:
        tuple: (X_train, X_test, y_train, y_test)
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)
    print(f"\nTraining data size (X_train, y_train): {X_train.shape}, {y_train.shape}")
    print(f"\nTraining data size (X_test, y_test): {X_test.shape}, {y_test.shape}")
    return X_train, X_test, y_train, y_test

# --- Building Model Neural Network ---

def build_neural_network_model(input_shape):
    """
    Builds a Sequential Neural Network model with several Dense and Dropout layers.

    Args:
        input_shape (int): Number of input features for the Neural Network model.

    Returns:
        tf.keras.Model: The compiled Neural Network model.
    """
    model = Sequential([
        Dense(128, activation='relu', input_shape=(input_shape,)), # Input layer with 128 neuron
        Dropout(0.3), # Dropout to prevent overfiting
        Dense(64, activation='relu'), # first hidden layer
        Dropout(0.3),
        Dense(32, activation='relu'), # second hidden layer
        # Sigmoid funtion for Binary classification
        Dense(1, activation='sigmoid') # Output layer for binary classification (0 or 1)
    ])

    # Compile the model using Adam optimizer and binary_crossentropy loss
    model.compile(optimizer=Adam(learning_rate=0.001),
                  loss='binary_crossentropy',
                  metrics=['accuracy'])

    print("\nNeural Network Model Summary:")
    model.summary()
    return model

# --- Training Model ---

def train_model(model, X_train_transformed, y_train, epochs=100, batch_size=32, validation_split=0.2):
    """
    Trains the Neural Network model and displays training history plots.

    Args:
        model (tf.keras.Model): The Neural Network model.
        X_train_transformed (np.array): The training feature data (after preprocessing).
        y_train (pd.Series): The training target data.
        epochs (int): Number of training epochs.
        batch_size (int): Batch size for training.
        validation_split (float): Proportion of training data to use for validation.

    Returns:
        tf.keras.callbacks.History: Training history object.
    """
    print("\nStarting Training model...")
    history = model.fit(X_train_transformed, y_train,
                        # epochs: how many times to go over the entire training set
                        epochs=epochs,
                        # batch_size: how many samples per mini-batch
                        batch_size=batch_size,
                        # 30% of training data is used for validation
                        validation_split=validation_split,
                        verbose=1) # verbose=1 menampilkan progress bar selama pelatihan

    # Plotting training and validation accuracy/loss history
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Training Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return history

# --- Evaluating Model ---

def evaluate_model(model, X_test_transformed, y_test):
    """
    Evaluates the model's performance on the test data and displays various classification metrics.

    Args:
        model (tf.keras.Model): The trained Neural Network model.
        X_test_transformed (np.array): The preprocessed test feature data.
        y_test (pd.Series): The test target labels.
    """
    print("\nEvaluating the model on test data:")
    loss, accuracy = model.evaluate(X_test_transformed, y_test, verbose=0)
    print(f"Test loss: {loss:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")

    # Mendapatkan probabilitas prediksi dan konversi ke label biner
    y_pred_proba = model.predict(X_test_transformed).ravel()
    y_pred = (y_pred_proba > 0.5).astype(int) # Ambang batas 0.5 untuk klasifikasi biner

    print("\nClassification Metrics:")
    # (True Positives/(TP + FP))
    print(f"Precision: {precision_score(y_test, y_pred):.4f}")
    # Recall (True Positives / (TP + FN))
    print(f"Recall: {recall_score(y_test, y_pred):.4f}")
    # Harmonic mean of Precision and Recall
    print(f"F1-Score: {f1_score(y_test, y_pred):.4f}")
    # Area Under the Receiver Operating Characteristic Curve
    print(f"AUC-ROC: {roc_auc_score(y_test, y_pred_proba):.4f}")

    print("\nConfusion Matrix:")
    cm = confusion_matrix(y_test, y_pred)
    # Visualising Confusion Matrix
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=[' No Disease', 'Has Disease'],
                yticklabels=[' No Disease', 'Has Disease'])
    plt.xlabel('Preddiction')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()

# --- Main Execution ---

if __name__ == "__main__":
    filepath = '/kaggle/input/heart-disease-cleveland-uci/heart_cleveland_upload.csv'

    target_column = None

    print("--- Starting data processing and NN model training---")

    # 1. load Data
    df = load_data(filepath)
    if df is None:
        exit() # there is no data to be loaded

    # --- VERIFIYING TARGET COLUMN ---
    print("\n")
    print("-------------------------------------------------------------------")
    print("IMPORTANT: Check the list of columns below to find the correct target column name.")
    print("The available columns in your dataset are:")
    print(df.columns.tolist())
    print("-------------------------------------------------------------------")
    print("\n")


    if 'condition' in df.columns:
        target_column = 'condition'
        print(f"Automatically selected 'condition' as the target column.")
    elif 'target' in df.columns:
        target_column = 'target'
        print(f"Automatically selected 'target' as the target column")
    elif 'HeartDisease' in df.columns:
        target_column = 'HeartDisease'
        print(f"Automatically selected 'HeartDisease' as the target column")
    elif 'output' in df.columns:
        target_column = 'output'
        print(f"Automatically selected 'output' as the target column")
    else:
        print("WARNING: Target column was not automatically detected")
        print("Please replace `target_column = None` with the correct target column name manually.")
        exit() #

    # Memastikan kolom target hanya berisi 0 dan 1.
    # Jika nilainya 0-4, perlu mengubahnya menjadi biner (0 untuk sehat, 1 untuk sakit)
    if df[target_column].max() > 1:
        print(f"\nTransforming'{target_column}' into binary (0: no disease, 1: disease).")
        df[target_column] = df[target_column].apply(lambda x: 1 if x > 0 else 0)
        print(f"Target column distribution after transformation: \n{df[target_column].value_counts()}")

    # 2. cleaning Data
    _ = check_missing_values(df)
    df_cleaned = clean_data(df)

    # inspect data
    explore_target_distribution(df_cleaned, target_column)

    # 3. preprocessing
    X, y, preprocessor, numerical_cols, categorical_cols = preprocess_features(df_cleaned, target_column)

    # 4. spliting Data
    X_train, X_test, y_train, y_test = split_data(X, y)

    # Applies scaling + encoding using ColumnTransformer
    X_train_transformed = preprocessor.fit_transform(X_train)
    X_test_transformed = preprocessor.transform(X_test)

    # 5. build model
    input_shape = X_train_transformed.shape[1]
    print(f"\nInput feature size after preprocessing: {input_shape}")

    model = build_neural_network_model(input_shape)

    # 6. train Model
    history = train_model(model, X_train_transformed, y_train, epochs=200, batch_size=32, validation_split=0.2)

    # 7. Evaluating Model
    evaluate_model(model, X_test_transformed, y_test)

    print("\n--- Data processing and NN model training completed---")
```
